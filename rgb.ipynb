```python
import matplotlib.pyplot as plt
from skimage import io #pedrito
from skimage import transform #petrolino

imagen_color = io.imread("perrito.jpg")

canal_rojo = imagen_color.copy ()
canal_rojo [:,:,1]= 0 #el 1 significa verde
canal_rojo [:,:,2]= 0 #el 2 significa azul

canal_azul = imagen_color.copy()
canal_azul [:,:,1]= 0
canal_azul [:,:,0]= 0

canal_verde = imagen_color.copy()
canal_verde [:,:,0]= 0
canal_verde [:,:,2]= 0

plt.figure(figsize=(15,9))
plt.subplot(221)
plt.imshow(imagen_color)
plt.title("image original")

plt.subplot(222)
plt.imshow(canal_rojo)
plt.title("imagen a color rojo")

plt.subplot(223)
plt.imshow(canal_azul)
plt.title("imagen a color azul")

plt.subplot(224)
plt.imshow(canal_verde)
plt.title("imagen a color verde")
```

```python
imagen_gris= io.imread("perrito.jpg", as_gray= True)

plt.figure(figsize=(15,9))
plt.subplot(221)
plt.imshow(imagen_gris)
plt.title("image original")

imagen_zoom_1 = imagen_gris[30:178,75:250]
plt.subplot(222)
plt.imshow(imagen_zoom_1)
plt.title("imagen zooom 1")
```
